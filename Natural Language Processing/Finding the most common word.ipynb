```python
import nltk
import requests
from bs4 import BeautifulSoup as bs
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
```

# Natural Language Processing
- find the ten most common meaningful words in the text, which ones are character names?

```python
#1. request the data
url = "https://www.gutenberg.org/files/16/16-h/16-h.htm"
req = requests.get(url)
req.encoding = "utf-8"
html = req.text


```

```python
#2. use BeautifulSoup to extract text
soup = bs(html,"html.parser")
text = soup.get_text().strip()
```

```python
##strip off the head notes on the top
delimiter = "Peter Pan, by J. M. Barrie"

text2 = text.split(delimiter)[1]

print(text2)
```

```python
#3. tokenize the data using NLTK
tokenizer = nltk.tokenize.RegexpTokenizer("\w+") #\w+ any words include digits, [A-Za-z] in the range A-Z or a-z
tokenized_word = tokenizer.tokenize(text2.lower())
print(tokenized_word)
```

```python
#4.lemmatize the doc
"""lemmitize = return the words to its root form"""
from nltk.stem import WordNetLemmatizer 

lemmatizer = WordNetLemmatizer()
lemma = [lemmatizer.lemmatize(word) for word in tokenized_word]
print(lemma[:30])
```

```python
#5. remove the non-alphabetic characters
a_lemma = [l for l in lemma if l.isalpha() ]
print(a_lemma) #you dont see any numerical values
```

```python
#6. Remove stop words from the tokenized word list
stop_word = stopwords.words("english")
print(stop_word[:10])

tokenized_nostop = [word for word in a_lemma if word not in stop_word]
print(tokenized_nostop[:10])
```

```python
#7. find the 10 most common words
#method 1
count = nltk.FreqDist(tokenized_nostop)
print(count.most_common(10))

#method 2
from collections import Counter
counters = Counter(tokenized_nostop)
print(counters.most_common(10))
```

```python
#8. find the 10 most common character names
tag = nltk.pos_tag(tokenized_nostop)
print(tag)
```

```python
#do a pair
word_tag_pairs = nltk.bigrams(tag)

#get a list of nouns
noun_preceders = [b[0] for (a, b) in word_tag_pairs if b[1] == 'NN']

#get the most frequent 10 character name
character_10_most_freq = [noun for noun in nltk.FreqDist(noun_preceders).most_common(10)]
```
